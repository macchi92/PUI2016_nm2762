CLEANING THE DATA AND MERGING THEM WITH THE GEOFILE

```python
def great_NM_HH_function (json):
    df=pd.DataFrame()
    df['geometry']=[]
    df['PO_NAME']=[]
    df['Shape_Leng']=[]
    df['Shape_Area']=[]
    df['borough']=[]
    df['postalCode']=[]
    
    for i in range(262):
        df = df.append({'geometry': Polygon(data['features'][i]['geometry']['coordinates'][0]),
                        'PO_NAME' : data['features'][i]['properties']['PO_NAME'],
                        'Shape_Area' : data['features'][i]['properties']['Shape_Area'],
                        'Shape_Leng' : data['features'][i]['properties']['Shape_Leng'],
                        'borough' : data['features'][i]['properties']['borough'],
                        'postalCode' : data['features'][i]['properties']['postalCode']
                       },ignore_index=True)

    return df
```

```python
json_to_geopandas=great_NM_HH_function(data['features'])
```

```python
json_to_geopandas=gpd.GeoDataFrame(json_to_geopandas)
```

```python
json_to_geopandas.columns=['geometry', 'PO_NAME', 'Shape_Leng', 'Shape_Area', 'borough',
       'zip']
```
